## Q4 Loyalty flag for existing customers

```python
ords_prods_merge_derived['max_order'] = ords_prods_merge_derived.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge_derived.head()
```

```python
ords_prods_merge_derived.loc[ords_prods_merge_derived['max_order']>40,'loyalty_flag']= 'Loyal customer'
```

```python
ords_prods_merge_derived.loc[(ords_prods_merge_derived['max_order']<=40)&(ords_prods_merge_derived['max_order']>10),'loyalty_flag']= 'Regular customer'
```

```python
ords_prods_merge_derived.loc[ords_prods_merge_derived['max_order']<=10,'loyalty_flag']= 'New customer'
```

```python
ords_prods_merge_derived['loyalty_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge_derived[['user_id', 'loyalty_flag', 'order_number','max_order']].head(60)
```

## Q6  Spending flag for each user based on the average price

```python
ords_prods_merge_derived['average_spending'] = ords_prods_merge_derived.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge_derived.head()
```

```python
ords_prods_merge_derived.loc[ords_prods_merge_derived['average_spending']<10,'spending_flag']= 'Low spender'
```

```python
ords_prods_merge_derived.loc[ords_prods_merge_derived['average_spending']>=10,'spending_flag']= 'High spender'
```

```python
ords_prods_merge_derived['spending_flag'].value_counts(dropna=False)
```

## Q7 Frequency flag acc to median in “days_since_prior_order” column

```python
ords_prods_merge_derived['median_days_since_prior_order'] = ords_prods_merge_derived.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge_derived.head()
```

```python
ords_prods_merge_derived.loc[ords_prods_merge_derived['median_days_since_prior_order']<=10,'frequency_flag']= 'Frequent customer'
```

```python
ords_prods_merge_derived.loc[(ords_prods_merge_derived['median_days_since_prior_order']>10) & (ords_prods_merge_derived['median_days_since_prior_order']<=20),'frequency_flag']= 'Regular customer'
```

```python
ords_prods_merge_derived.loc[ords_prods_merge_derived['median_days_since_prior_order']>20,'frequency_flag']= 'Non-frequent customer'
```

```python
ords_prods_merge_derived['frequency_flag'].value_counts(dropna=False)
```
