# User Defined Function using if-statements

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range']= df.apply(price_label,axis=1)
```

```python
df['price_range'].value_counts(dropna=False)
```

# Q2- Creating column for busiest days (orders_day_of_week=0 or 1), slowest days (orders_day_of_week=3 or 4), regularly busy(orders_day_of_week=2or 5 or 6) using if statememts with for loop.

```python
result_1 = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_1.append("Busiest days")
  elif value == 4 or value == 3:
    result_1.append("Slowest days")
  else:
    result_1.append("Regularly busy")
```

```python
result_1
```

# Creating Busiest_days column 

```python
ords_prods_merge['busiest_days']=result_1
```
